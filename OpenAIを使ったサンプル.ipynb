```python
# モジュールのインポート
import os
import base64
from PIL import Image
import io
import openai

# colabに設定した環境変数を取得するライブラリ
from google.colab import userdata
```

```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

# DALL・E 3を使った画像生成
本書(ソロマメ)の表紙に使ったプロンプトを利用して画像生成を試してみます。

```python
response = client.images.generate(
    # 利用するモデル
    model="dall-e-3",
    # プロンプト
    prompt="masterpiece, best quality, solo, a very cute and beautiful anime girl ,very high quality clear anime amber eyes, sky blue hair, twin tails, hair ribbon, upper body right side standing, school uniform, cardigan, watercolor, autumn leaves background, golden sunlight filtering through the trees, single character",
    # 画像サイズ
    size="1024x1024",
    # 画像の品質(DALL・E 3 standard or hd)
    quality="standard",
    # 生成する画像数(DALL・E 3の場合は1のみ)
    n=1,
    # 出力フォーマット
    response_format="b64_json"
)

print(response)
```

```python
# 画像のURLの取得(ここをクリックすると画像を閲覧できます)
print(response.data[0].url)
```

```python
# 出力された画像を保存
encoded_data = response.data[0].b64_json
decoded_data = base64.b64decode(encoded_data)
image = Image.open(io.BytesIO(decoded_data))

# 画像ファイルとして保存
image.save('output_image.png')
```

```python
# 画像の表示
image
```

# GPT-4 with Visionで画像からテキストを生成
次にGPT-4vを使って画像からテキストの生成を試してみます  
入力する画像として先ほど生成した画像を利用して画像の説明を生成してみましょう

```python
# 画像の質問応答
response = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "これは何の画像ですか？"},
                {
                    "type": "image_url",
                    "image_url": f"data:image/jpeg;base64,{encoded_data}",
                },
            ],
        }
    ],
    max_tokens=300,
)

print(response.choices[0].message.content)
```
